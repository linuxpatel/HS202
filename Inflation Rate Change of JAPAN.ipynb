```python
# importing the required module
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# AVG INFLATION IN YEAR
x = [2021,2020,2019,2018,2017,2016,2015]
# INFLATION 
y = [-0.49,-0.01,0.48,0.98,0.47,-0.12,0.80]
df = pd.DataFrame(list(zip(x, y)),
               columns =['Year', 'Avf_Inf'])
df
```

```python
#PLOT OF AVERAGE INFLATION JAPAN(CPI) - BY YEAR

# importing the required module
import matplotlib.pyplot as plt
  
# AVG INFLATION IN YEAR
x = [2021,2020,2019,2018,2017,2016,2015]
# INFLATION 
y = [-0.49,-0.01,0.48,0.98,0.47,-0.12,0.80]
  
# plotting the points 
plt.plot(x, y,marker = 'o')
  
# naming the x axis
plt.xlabel('Year')
# naming the y axis
plt.ylabel('Average Inflation (CPI) - by year')
plt.grid(True)  
# giving a title to my graph
plt.title('Average Inflation Chart of Japan (CPI) year to year basis ')
 
# function to show the plot
plt.show()
```

```python
# plotting strip plot with seaborn
ax = sns.stripplot(x, y);
 
# giving labels to x-axis and y-axis
ax.set(xlabel ='Year', ylabel ='Avg_Inflation')
sns.barplot(x, y, data = df, palette ='plasma')
# giving title to the plot
plt.title('Average Inflation Chart of Japan (CPI) year to year basis ')
 
# function to show plot
plt.show()
```

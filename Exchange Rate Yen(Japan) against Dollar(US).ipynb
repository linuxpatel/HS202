```python
# importing the required module
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# AVG INFLATION IN YEAR
x = ['Feb 2021','Jan 2021','Dec 2020','Nov 2020','Oct 2020','Sep 2020','Aug 2020','Jul 2020','Jun 2020','May 2020','Apr 2020','Mar 2020']
# INFLATION 
x.reverse()
y = [74.111,75.827,76.068,77.032,77.694,78.076,78.306,78.708,78.778,80.615,80.264,79.109]
y.reverse()
df = pd.DataFrame(list(zip(x, y)),
               columns =['Year', 'Avf_Inf'])
df
```

```python
#PLOT OF AVERAGE INFLATION JAPAN(CPI) - BY YEAR

# importing the required module
import matplotlib.pyplot as plt
  
  
# plotting the points 
plt.plot(x, y,marker = 'o')
  
# naming the x axis
plt.xlabel('Year')
# naming the y axis
plt.ylabel('Average Inflation (CPI) - by year')
plt.grid(True)  
# giving a title to my graph
plt.title('Average Inflation Chart of Japan (CPI) year to year basis ')
# Rotating X-axis labels
plt.xticks(rotation = 45)

# function to show the plot
plt.show()
```

```python
# plotting strip plot with seaborn
ax = sns.stripplot(x, y);
 
# giving labels to x-axis and y-axis
ax.set(xlabel ='Year', ylabel ='Avg_Inflation')
sns.barplot(x, y, data = df, palette ='plasma')
# giving title to the plot
plt.title('Average Inflation Chart of Japan (CPI) year to year basis ')
# Rotating X-axis labels
plt.xticks(rotation = 45)
   
# function to show plot
plt.show()
```
